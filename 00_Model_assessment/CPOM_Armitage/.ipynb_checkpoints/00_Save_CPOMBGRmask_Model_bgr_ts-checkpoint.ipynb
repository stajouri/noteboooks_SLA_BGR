# loading CPOM arctic data

```python
chunk_size = {'lat':25,'lon':80}
diri = '/gpfswork/rech/cli/uor98hu/DATA4VALIDATION/Armitage/'
namo = 'CPOM_DOT.nc'
CPOMdat = xr.open_dataset(diri+namo,chunks=chunk_size)
CPOMdata = CPOMdat.rename({'date':'time_counter','lon':'x','lat':'y'})
# 2003-01 to 2014-12: change time unit
time = pd.date_range(start='01/2003', end='01/2015', freq='M')
# create mesh grid as coordinates
xx,yy = np.meshgrid(np.unique(CPOMdata.x),np.unique(CPOMdata.y))
longs =xr.DataArray(xx, dims=('y','x'))
lats  =xr.DataArray(yy, dims=('y','x'))
CPOMdata = CPOMdata.assign_coords(nav_lon=longs, nav_lat=lats, time_counter=time)

DOT = CPOMdata.DOT
```

## interpolating the BGRmask of the model onto the CPOM product grid

### interpolating on the whole grid and save the mask

```python
#### model 

BGRmask = xr.open_dataset('/gpfswork/rech/cli/uor98hu/MYDATA/TRANSPORT_WP1/BGRmaskobs.nc').BGRmask.squeeze()

dirmask="/gpfswork/rech/cli/rcli002/eORCA025.L75/eORCA025.L75-I/"
mesh_hgr = xr.open_dataset(dirmask+'mesh_hgr.nc').squeeze().rename({"nav_lev":"deptht"}).reset_coords(names='time_counter', drop=True)

nav_lon =  mesh_hgr.nav_lon.fillna(0)
nav_lat =  mesh_hgr.nav_lat.fillna(0)

model_grid = (nav_lat.values.ravel(), nav_lon.values.ravel() )

BGRmask = BGRmask.assign_coords(nav_lon=nav_lon, nav_lat=nav_lat)
```

```python
### apply interpolation
# scipy.interpolate.griddata
grid_z0 = griddata(model_grid, BGRmask.values.ravel(), (lats.values, longs.values), method='nearest')
grid_z0[84,:] = 0

CPOMmaskfull = xr.DataArray(
    data=grid_z0,
    dims=["y", "x"],
    coords=dict(
        nav_lon=(["y", "x"], longs.data),
        nav_lat=(["y", "x"], lats.data)))
dsCPOMmaskfull = CPOMmaskfull.to_dataset(name='BGRmaskCPOM')
```

```python
dsCPOMmaskfull.to_netcdf("/gpfswork/rech/cli/uor98hu/DATA4VALIDATION/Armitage/CPOM_DOT_BGRmask.nc", mode='w') 
```

# for model data : saving the time series 1980-2018 of the mean yearly/monhtly ssh in the BGR box

```python
# mask data
pathmask = '/gpfswork/rech/cli/uor98hu/MYDATA/TRANSPORT_WP1/'
maskds = xr.open_dataset(pathmask+'BGRmaskobs.nc').squeeze()
BGRmask = maskds.BGRmask
```

```python
dirmask="/gpfswork/rech/cli/rcli002/eORCA025.L75/eORCA025.L75-I/"
mesh_hgr = xr.open_dataset(dirmask+'mesh_hgr.nc').squeeze().rename({"nav_lev":"deptht"})
tmask = mesh_hgr.tmask

e1t = mesh_hgr.e1t.fillna(0)
e2t = mesh_hgr.e2t.fillna(0)

nav_lon =  mesh_hgr.nav_lon
nav_lat =  mesh_hgr.nav_lat

# constructing the weight
bt = (e1t * e2t).load()# area of each cell
btboxBG = bt.where(BGRmask,drop=True)
tmaskboxBG = tmask.where(BGRmask,drop=True)
oceanArea = (btboxBG*tmaskboxBG[0])  #tmask : 1 on ocean, 0 on land
totalOceanArea = oceanArea.sum()
WeightboxBG = (oceanArea/totalOceanArea) #.compute()
```

```python
WeightboxBG.plot()
WeightboxBG.sum().values
```

```python
# ssh data
pathww = '/gpfswork/rech/cli/uor98hu/MYDATA/SSH/GAI/'
ds = xr.open_dataset(pathww+'eORCA025.L75-IMHOTEP.GAI_y1980_2018.1m_sossheigCorrected.nc')
ssh = ds.sossheig
```

```python
#selecting the area of interest
# ssh_zone = (SSHcorrgmsl1y.where(BGRmask,drop=True) * WeightboxBG)

#  montlhy
ssh_zone = (ssh.where(BGRmask,drop=True) * WeightboxBG)
```

```python
ssh_zone_mean = ssh_zone.sum(('x','y'))
sssh_zone_final = ssh_zone_mean - ssh_zone_mean.mean()
```

```python
(sssh_zone_final*100).plot()
```

```python
sssh_zone_final = sssh_zone_final.assign_attrs({"long_name":"GAI ssh corrected from GMSL of GAI en monthly frequency, the space average has been weighted by the local weight gridcell area in the the BGRbox area defined using BGRmaskobs.nc"})
```

```python
final_ds = sssh_zone_final.to_dataset(name='ssh_BGRbox',promote_attrs=True)
```

```python
final_ds.to_netcdf(path="/gpfswork/rech/cli/uor98hu/MYDATA/SSH/GAI/GAI_ts1980-2018_ssh_BGRbox_1m.nc",mode="w")
```
