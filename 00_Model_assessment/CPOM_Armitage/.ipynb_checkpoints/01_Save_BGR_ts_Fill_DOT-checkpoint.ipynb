# loading CPOM arctic data

```python
# first manipulations of the dataset for ease: coordinates
chunk_size = {'lat':25,'lon':80}
diri = '/gpfswork/rech/cli/uor98hu/DATA4VALIDATION/Armitage/'
namo = 'Full_DOT_data_Arco.nc'
CPOMdat = xr.open_dataset(diri+namo,chunks=chunk_size)
# 2011-01 to 2020-12: change time unit
time = pd.date_range(start='01/2011', end='01/2021', freq='M')
CPOMdata = CPOMdat.rename({'time':'time_counter'})
CPOMnav_lat = CPOMdata.lats
CPOMnav_lon = CPOMdata.lons
CPOMdata = CPOMdata.assign_coords(nav_lon=CPOMnav_lon, nav_lat=CPOMnav_lat, time_counter=time)
sla = CPOMdata.Sea_level_anomaly
```

# Apply the BGRmaskCPOM to CPOM sla  and do the time series on the box

```python
BGRmaskCPOM = xr.open_dataset('/gpfswork/rech/cli/uor98hu/DATA4VALIDATION/Armitage/Full_DOT_data_Arco_BGRmask.nc').BGRmaskCPOM.squeeze()

sla_masked = sla.where(BGRmaskCPOM,drop=True).sel(time_counter=slice('2011','2018'))
sla_masked_1y = sla_masked.groupby('time_counter.year').mean('time_counter')
```

```python
xdist = CPOMdata.xdist
ydist = CPOMdata.ydist
cellarea = xdist*ydist
cellareaBGr = cellarea.where(BGRmaskCPOM,drop=True)
totalOceanArea = cellareaBGr.sum()
WeightBGR_CPOM = (cellareaBGr/totalOceanArea) #.compute()
```

```python
# weighted mean in the BGRbox
sla_boxmasked_mean_1m = (sla_masked*WeightBGR_CPOM).sum(('x','y'))
sla_boxmasked_mean_1m = sla_boxmasked_mean_1m - sla_boxmasked_mean_1m.mean()

sla_boxmasked_mean_1y = (sla_masked_1y*WeightBGR_CPOM).sum(('x','y'))
sla_boxmasked_mean_1y = sla_boxmasked_mean_1y - sla_boxmasked_mean_1y.mean()
```

# compute the trend 

```python
def DTUtrend1m(DTU,dec_year,t1,t2):
    """
    Arguments : 
    DTU : DTU arctic data as xarray contains the variables date and sea_level_anomaly
    t1 : start date in the format "year-mm" of the period
    t2 : end   date in the format "year-mm" of the period
    """
    #######------- PREPROCESSING
    #cropping the data in time to get the period of interest
    ## of DTU data
    DTUper = DTU.sel(time_counter = slice(t1,t2))
    dec_yearper = dec_year.sel(time_counter = slice(t1,t2))

    ########------- Calculating the linear regression
    
    # 1/ changing the time coordinate to do the linear regression
    DTUper = DTUper.assign_coords(time_counter = dec_yearper)
    # 2/ do the linear regression
    SLAfit = DTUper.polyfit(dim='time_counter', deg=1).compute()
    SLAtrend = SLAfit.polyfit_coefficients[0]
    SLAintercept = SLAfit.polyfit_coefficients[1]

    return SLAtrend, SLAintercept, dec_yearper
```

```python
# trend 1m

#period of interest
t1 = "2005-01"
t2 = "2018-09"
time4plt = sla_boxmasked_mean_1m.sel(time_counter = slice(t1,t2)).time_counter.values
dec_year = DTUdata.date

#-------------- DTU product
SLAtr, SLAintercept, dec_yearper = DTUtrend1m((sla_boxmasked_mean_1m),dec_year,t1,t2)
### trend curve
tr_sla_boxmasked_mean_1m = SLAtr.values*dec_yearper.values + SLAintercept.values

#-------------- model

SSHtr, SSHintercept, dec_yearper = DTUtrend1m((ssh_1m),dec_year,t1,t2)
### trend curve
tr_ssh_1m = SSHtr.values*dec_yearper.values + SSHintercept.values
```
