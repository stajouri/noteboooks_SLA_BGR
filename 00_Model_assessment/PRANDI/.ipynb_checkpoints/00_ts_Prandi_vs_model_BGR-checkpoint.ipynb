## Apply the BGRmaskAlti to sla  and do the time series on the box

```python
BGRmaskD = xr.open_dataset('/gpfswork/rech/cli/uor98hu/DATA4VALIDATION/PRANDI/PRANDI_BGRmask.nc').BGRmaskPrandi.squeeze()
```

```python
# weight of this grid
cellArea = 25e3*25e3
matrix_cellArea = np.ones((350,350))*cellArea

da_cell_area = xr.DataArray(
    data=matrix_cellArea,
    dims=["x", "y"],
    coords=dict(
        nav_lon=(["x", "y"], PRANDIdata.nav_lon.data),
        nav_lat=(["x", "y"], PRANDIdata.nav_lat.data)))

da_cell_area_mskd = da_cell_area.where(BGRmaskD)
```

```python
mask_cellArea_tot = da_cell_area_mskd.sum()
da_weight4_mskd = da_cell_area_mskd /mask_cellArea_tot
```

```python
# da_weight4_mskd.sum()
da_weight4_mskd
sla_boxmasked
```

```python
sla_boxmasked = sla.where(BGRmaskD)
sla_boxmasked_mean_1m = (sla_boxmasked*da_weight4_mskd).sum(('y','x'))
sla_boxmasked_mean_1m = sla_boxmasked_mean_1m - sla_boxmasked_mean_1m.mean()

# sla_boxmasked_1y = sla_boxmasked.groupby('time_counter.year').mean('time_counter')
```
