# loading DTU/TUM arctic data

```python
chunk_size = {'lat':25,'lon':80}
diri = '/gpfswork/rech/cli/uor98hu/DATA4VALIDATION/'
namo = 'cci_dtu_tum_arctic_sla_v3.1.nc'
DTUdat = xr.open_dataset(diri+namo,chunks=chunk_size)
#sept 1991 to sept 2018: change time unit
time = pd.date_range(start='09/1991', end='10/2018', freq='M')
DTUdata = DTUdat.rename({'t':'time_counter', 'lat':'y', 'lon':'x'})
# create mesh grid as coordinates
xx,yy = np.meshgrid(np.unique(DTUdata.longitude),np.unique(DTUdata.latitude))
longs =xr.DataArray(xx, dims=('y','x'))
lats =xr.DataArray(yy, dims=('y','x'))
DTUdata = DTUdata.assign_coords(nav_lon=longs, nav_lat=lats, time_counter=time)
```

## interpolating the BGRmask of the model onto the DTU product grid

### interpolating on a smaller, same region

```python
##### model 

BGRmask = xr.open_dataset('/gpfswork/rech/cli/uor98hu/MYDATA/TRANSPORT_WP1/BGRmaskobs.nc').BGRmask.squeeze()

dirmask="/gpfswork/rech/cli/rcli002/eORCA025.L75/eORCA025.L75-I/"
mesh_hgr = xr.open_dataset(dirmask+'mesh_hgr.nc').squeeze().rename({"nav_lev":"deptht"}).reset_coords(names='time_counter', drop=True)

nav_lon =  mesh_hgr.nav_lon.fillna(0)
nav_lat =  mesh_hgr.nav_lat.fillna(0)

nav_lon_box =  nav_lon.where((nav_lon<-130) & (nav_lon>-170) & (nav_lat<81) & (nav_lat>70), drop=True).fillna(0) # have nan values
nav_lat_box =  nav_lat.where((nav_lon<-130) & (nav_lon>-170) & (nav_lat<81) & (nav_lat>70), drop=True).fillna(0) # have nan values

model_grid = (nav_lat_box.values.ravel(), nav_lon_box.values.ravel() )

mask_box =  BGRmask.where((nav_lon<-130) & (nav_lon>-170) & (nav_lat<81) & (nav_lat>70), drop=True).fillna(0)
mask_box = mask_box.assign_coords(nav_lon=nav_lon_box, nav_lat=nav_lat_box)
```

```python
### DTU product
lats_box = lats.where((longs<-130) & (longs>-170) & (lats<81) & (lats>70),drop=True).fillna(0)
longs_box = longs.where((longs<-130) & (longs>-170) & (lats<81) & (lats>70),drop=True).fillna(0)
```

```python
### apply interpolation
# scipy.interpolate.griddata

grid_z0 = griddata(model_grid, mask_box.values.ravel(), (lats_box.values,longs_box.values), method='nearest')

DTUmask = xr.DataArray(
    data=grid_z0,
    dims=["y", "x"],
    coords=dict(
        nav_lon=(["y", "x"], longs_box.data),
        nav_lat=(["y", "x"], lats_box.data)))

dsDTUmask = DTUmask.to_dataset(name='BGRmaskDTU')
```
