```python
# ------------------------
## ----import libraries
# ------------------------
import os,sys
import numpy as np
# xarray
import xarray as xr
import pandas as pd

# plot
import matplotlib.pyplot as plt
import matplotlib.image as image
from matplotlib.offsetbox import OffsetImage,AnchoredOffsetbox
```

```python
file_in = '/gpfswork/rech/cli/uor98hu/MYDATA/nao.dat'
file_out = '/gpfswork/rech/cli/uor98hu/MYDATA/nao4memonthly.dat'
```

```python
fin = open(file_in, "rt")
fout = open(file_out, "wt")
i = 0
for line in fin:
    l = line.split()
    #modify first line
    if i == 0 :  l = ['Year','01','02','03','04','05','06','07','08','09','10','11','12','Annual']
    fout.write(' '.join(l)+'\n')
    i+=1
    
fin.close()
fout.close()
```

```python
data = pd.read_csv(file_out,sep=" ")
```

# Annual NAO 

```python
time = data.loc[158:196,'Year']
NAO = data.loc[158:196,'Annual']
```

# monthly NAO

```python
lmonth = ['01','02','03','04','05','06','07','08','09','10','11','12']
path = "/gpfsstore/rech/cli/rcli002/eORCA025.L75/eORCA025.L75-IMHOTEP.GAI-S/1m/"
year = 1980

for i in np.arange(158,197):
    datadate = xr.open_mfdataset(path+str(year)+"/eORCA025.L75-IMHOTEP.GAI_y"+str(year)+"m*.1m_gridTsurf.nc")
    
    NAOm = data.loc[i,lmonth]
    NAOds = xr.Dataset(
        data_vars=dict(NAO =(["time_counter"],NAOm)),
        coords=dict(time_counter=datadate.time_counter.values))
    namo="NAOm"+str(year)+".nc"
    NAOds.to_netcdf(path="/gpfswork/rech/cli/uor98hu/MYDATA/NAO/"+namo, mode='w')
    year+=1
```
