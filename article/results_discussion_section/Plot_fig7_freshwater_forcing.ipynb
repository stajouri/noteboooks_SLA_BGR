```python
def place_image(im, loc=3, ax=None, zoom=1, **kw):
    if ax==None: ax=plt.gca()
    imagebox = OffsetImage(im, zoom=zoom*0.72)
    ab = AnchoredOffsetbox(loc=loc, child=imagebox, frameon=False, **kw)
    ax.add_artist(ab)
```

# Load data: GREENLAND

```python
# load GAI data
Grall_perbasin_GAI = xr.open_dataset('/gpfswork/rech/cli/uor98hu/MYDATA/Gr_contribution/Gr_forcing_verif/Gr_mean_alltypes_perbasin_GAI.nc')

totperyrGAI = Grall_perbasin_GAI.Greenland #Grall_perbasin_GAI.Southwest + Grall_perbasin_GAI.Central_West + Grall_perbasin_GAI.Northwest + \
                # Grall_perbasin_GAI.North     + Grall_perbasin_GAI.Northeast    + Grall_perbasin_GAI.Central_east + \
                # Grall_perbasin_GAI.southeast

# load AI data
Grall_perbasin_AI = xr.open_dataset('/gpfswork/rech/cli/uor98hu/MYDATA/Gr_contribution/Gr_forcing_verif/Gr_mean_alltypes_perbasin_AI.nc').roll(time_counter=-1)

# create array of 39 year of AI forcing per region
SWtotperyearAI = np.ones((39))*(Grall_perbasin_AI.Southwest.mean('time_counter')).values 
CWtotperyearAI = np.ones((39))*(Grall_perbasin_AI.Central_West.mean('time_counter')).values
NWtotperyearAI = np.ones((39))*(Grall_perbasin_AI.Northwest.mean('time_counter')).values
NtotperyearAI = np.ones((39))*(Grall_perbasin_AI.North.mean('time_counter')).values
NEtotperyearAI = np.ones((39))*(Grall_perbasin_AI.Northeast.mean('time_counter')).values 
CEtotperyearAI = np.ones((39))*(Grall_perbasin_AI.Central_east.mean('time_counter')).values 
SEtotperyearAI = np.ones((39))*(Grall_perbasin_AI.southeast.mean('time_counter')).values 
GrtotperyearAI = np.ones((39))*(Grall_perbasin_AI.Greenland.mean('time_counter')).values

# convert it into dataset
dsAI = xr.Dataset(
    data_vars=dict(Southwest=(["time_counter"],    SWtotperyearAI),
                   Central_West=(["time_counter"], CWtotperyearAI),
                   Northwest=(["time_counter"],    NWtotperyearAI),
                   North=(["time_counter"],        NtotperyearAI),
                   Northeast=(["time_counter"],    NEtotperyearAI),
                   Central_east=(["time_counter"], CEtotperyearAI),
                   southeast=(["time_counter"],    SEtotperyearAI),
                    Greenland=(["time_counter"],   GrtotperyearAI),),
    coords=dict(time_counter=Grall_perbasin_GAI.time_counter.values),
    attrs=dict(description="Greenland's total melting (0 to 10 m) per basin for run AI [mSv] "),
                )

# Do GAI- AI on annual data 
GAI_AI_Grforcing = Grall_perbasin_GAI - dsAI
```

# Load data: RIVERS

```python
# load data AI
rnf_perbasin_AI = xr.open_dataset('/gpfswork/rech/cli/uor98hu/MYDATA/rivers_mask/rivers_mean_rnf_perbasin_GAI.nc')

totperyrAI = rnf_perbasin_AI.chukchi + rnf_perbasin_AI.eastsiberian + rnf_perbasin_AI.laptev + rnf_perbasin_AI.kara + rnf_perbasin_AI.barents.data \
            + rnf_perbasin_AI.norwegian + rnf_perbasin_AI.caa        + rnf_perbasin_AI.beaufort     + rnf_perbasin_AI.baffinHudson 

# load data S
rnf_perbasin_S = xr.open_dataset('/gpfswork/rech/cli/uor98hu/MYDATA/rivers_mask/rivers_mean_monthly_rnf_perbasin_S.nc')

# create array of 39 year of S forcing per region
SWtotperyearAI = np.ones((39))*(rnf_perbasin_S.chukchi.mean('time_counter')).values 
CWtotperyearAI = np.ones((39))*(rnf_perbasin_S.eastsiberian.mean('time_counter')).values
NWtotperyearAI = np.ones((39))*(rnf_perbasin_S.laptev.mean('time_counter')).values
NtotperyearAI  = np.ones((39))*(rnf_perbasin_S.kara.mean('time_counter')).values
NEtotperyearAI = np.ones((39))*(rnf_perbasin_S.barents.mean('time_counter')).values 
CEtotperyearAI = np.ones((39))*(rnf_perbasin_S.norwegian.mean('time_counter')).values 
GrtotperyearAI = np.ones((39))*(rnf_perbasin_S.baffinHudson.mean('time_counter')).values
SEtotperyearAI = np.ones((39))*(rnf_perbasin_S.caa.mean('time_counter')).values 
BetotperyearAI = np.ones((39))*(rnf_perbasin_S.beaufort.mean('time_counter')).values

# convert it into datase
dsS = xr.Dataset(
    data_vars=dict(chukchi =(["time_counter"],      SWtotperyearAI),
                   eastsiberian =(["time_counter"], CWtotperyearAI),
                   laptev =(["time_counter"],       NWtotperyearAI),
                   kara =(["time_counter"],         NtotperyearAI),
                   barents =(["time_counter"],      NEtotperyearAI),
                   norwegian =(["time_counter"],    CEtotperyearAI),
                   baffinHudson =(["time_counter"], GrtotperyearAI),
                   caa =(["time_counter"],          SEtotperyearAI),
                   beaufort =(["time_counter"],     BetotperyearAI),),

    coords=dict(time_counter=rnf_perbasin_AI.time_counter.values),
    attrs=dict(description="Rivers's total melting per basin for run S [mSv] "),
                )

# Do AI-S on annual data 
AI_S_Rivforcing = rnf_perbasin_AI - dsS
```

```python
RivTot = AI_S_Rivforcing.chukchi+\
        AI_S_Rivforcing.eastsiberian+\
        AI_S_Rivforcing.laptev+\
        AI_S_Rivforcing.kara+\
        AI_S_Rivforcing.barents+\
        AI_S_Rivforcing.norwegian+\
        AI_S_Rivforcing.baffinHudson+\
        AI_S_Rivforcing.caa+\
        AI_S_Rivforcing.beaufort

am_riv = AI_S_Rivforcing.chukchi +  AI_S_Rivforcing.beaufort + AI_S_Rivforcing.caa + AI_S_Rivforcing.baffinHudson
ru_riv = AI_S_Rivforcing.eastsiberian + AI_S_Rivforcing.laptev + AI_S_Rivforcing.kara+ AI_S_Rivforcing.barents + AI_S_Rivforcing.norwegian
```
