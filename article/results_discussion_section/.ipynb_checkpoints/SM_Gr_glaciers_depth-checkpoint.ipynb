```python
## standart libraries
import os,sys
import numpy as np

# xarray
import xarray as xr
# plot
import cartopy
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import matplotlib.pyplot as plt
from matplotlib.colors import Colormap
import matplotlib.cm as cm
```

```python
# read grid files
diri="/gpfswork/rech/cli/rcli002/eORCA025.L75/eORCA025.L75-I/"
nav_lon = xr.open_dataset(diri+'mesh_hgr.nc')['nav_lon'].squeeze()
nav_lat = xr.open_dataset(diri+'mesh_hgr.nc')['nav_lat'].squeeze()
```

```python
#vue polaire
loncentr=320 
latcentr=70
sath=2085831
```

```python
# These files combine runoff depths coming from ISBA and GrIS, and rnfisf_depht coming from Greenland and Antarctica (3 different sources). 
# DEPdata = xr.open_dataset(diri+'eORCA025.L75_rnf_dep.nc')
# rnf_min = DEPdata.sozisfmin
# rnf_max = DEPdata.sozisfmax
# rnf_mask = xr.open_dataset(diri+'eORCA025.L75_rnf_msk.nc').socoefr

# this file contain only data for Gr
dep_gr = xr.open_dataset(diri+'eORCA025.L75_y1950-2020_1m_greenland_isfbis.nc')
dep_gr_min = dep_gr.sozisfmin
dep_gr_max = dep_gr.sozisfmax
```

```python
#========= CREATE FIGURE
diro = "/gpfsscratch/rech/cli/uor98hu/PLOTS/"
namo  ="Greenland_SI_ISFdepths.png"
print(namo)
#===========================================
fig = plt.figure(figsize=([18,10]),facecolor='none', dpi=72*2)

trdata  = ccrs.PlateCarree() # Data system proj (if coords are in lat lon, use PlateCarre here)
ax = plt.axes(projection= ccrs.NearsidePerspective(central_longitude=loncentr, central_latitude=latcentr,satellite_height=sath))

extent = [-57,-25,58,84] # (x0, x1, y0, y1)
ax.set_extent(extent)

cl2 = ax.add_feature(cfeature.LAND.with_scale('50m'),facecolor="none",edgecolor='k',alpha=1, zorder=5, linewidth=0.5)

cs = ax.pcolormesh(nav_lon, nav_lat, dep_gr_max.where(dep_gr_max!=0),cmap =cm.Reds, transform=trdata)
              
cbar = fig.colorbar(cs)
cbar.ax.tick_params(labelsize=14)
cbar.set_label(r'deeper depth of the grounding line [m]', fontsize=20)

plt.show()
# fig.savefig(diro+namo, facecolor=fig.get_facecolor(), bbox_inches='tight',edgecolor='none') 
plt.close(fig)
```
