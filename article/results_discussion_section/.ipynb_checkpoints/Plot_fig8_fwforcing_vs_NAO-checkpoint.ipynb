## Detrending of GREENLAND time series

```python
yr_Gr_1y.plot()
```

```python
def detrend_dim(da, dim, deg=1):
    # detrend along a single dimension
    p = da.polyfit(dim=dim, deg=deg)
    fit = xr.polyval(da[dim], p.polyfit_coefficients)
    return da - fit

#-------------------------------------------------------
p = yr_Gr_1y.polyfit('time_counter', deg=2)
fit2 = xr.polyval(yr_Gr_1y.time_counter, p.polyfit_coefficients)
detrended2 = yr_Gr_1y - fit2


#--------------------------------------------------------
fig = plt.figure(figsize=(20,4), facecolor='none', dpi=72*2)

yr_Gr_1y.plot(linewidth=2)
fit2.plot(linewidth=2)
detrended2.plot(linewidth=2)

plt.legend(['initial','fit ordre 2','detrended2'])
plt.grid()
plt.xlabel('Time')

plt.show()
plt.close(fig)
```

```python
# test of significance of the correlations
r1 = 0.095
r2 = -0.42
r3 = -0.51
r4 = np.round(xr.corr(NAOmeanwinter, yr_Riv_RU1y).values,2)
r5 = np.round(xr.corr(NAOmeanwinter, yr_Riv_AM1y).values,2)

#compute the t value: n is the lenght of the time series
def compute_t(r, n):
    return r /np.sqrt((1-r**2)/(n-2))

t1 = compute_t(r1,39)
t2 = compute_t(abs(r2),38)
t3 = compute_t(abs(r3),39)
t4 = compute_t(abs(r4),39)
t5 = compute_t(abs(r5),39)

print(t1)
print(t2)
print(t3)
print(t4)
print(t5)
```

```python
from scipy import stats

# compute t* : (stats.t.ppf(1-alpha,dof-2))
print(stats.t.ppf(1-0.05,39))
print(stats.t.ppf(1-0.05,38))

# is significant if t is > t*
```
