```python
def do_bilan(nexpREF):
    # open data and read the variables of the budget

    dScumsum     = xr.open_dataset(dirww+nexpREF+"_dScumsum_1980_2018_after_review.nc") .dScumsum[:-1]
    adv_h_cumsum = xr.open_dataset(dirww+nexpREF+"_adv_h_cumsum_1980_2018_after_review.nc").adv_h_cumsum[:-1]
    sfx_cumsum   = xr.open_dataset(dirww+nexpREF+"_sfx_cumsum_1980_2018_after_review.nc").sfx_cumsum[:-1]
    # sfx_cumsum   = xr.open_dataset(dirww+"G_forcing/"+nexpREF+"_sfx_cumsum_1980_2018.nc").sfx_cumsum[:-1]

    # deduce the residual
    res_cumsum = dScumsum -(adv_h_cumsum + sfx_cumsum)
    

    return [dScumsum, adv_h_cumsum, sfx_cumsum, res_cumsum]
```

```python
GAIlist = do_bilan("GAI")
AIlist = do_bilan("AI")
Slist = do_bilan("S")

# GAI-S
GAI_Slist = [GAIlist[i]-Slist[i] for i in np.arange(4)]

# GAI-AI
GAI_AIlist = [GAIlist[i]-AIlist[i] for i in np.arange(4)]

# AI-S
AI_Slist = [AIlist[i]-Slist[i] for i in np.arange(4)]
```

```python
from scipy import stats
import esmtools
```

```python
# test of significance of correlation coefficients
#compute the t value: n is the lenght of the time series
def compute_t(r, n):
    return r /np.sqrt((1-r**2)/(n-2))
```

```python
## correlation between Gtotal GAI et SLA GAI
exp1_gtotal = (GAIlist[0]* 1e3)
expA_sla = xr.open_dataset('/gpfswork/rech/cli/uor98hu/MYDATA/SSH/GAI/GAI_ts1980-2018_ssh_BGRbox_1m.nc').ssh_BGRbox
r = np.round(xr.corr((GAIlist[0]* 1e3), expA_sla).values,2)
print(r)

t = compute_t(abs(r),len(exp1_gtotal))
print(t)
# compute t* : (stats.t.ppf(1-alpha,dof-2))
tstar = stats.t.ppf(1-0.05,len(exp1_gtotal)-2)
print(tstar)
print("is significant if true:", t>tstar)
# is significant if t is > t*
```

```python
# trend of gr+riv Gtotal and the significance
gai_s_gtotal = (GAI_Slist[0]* 1e3).groupby('time_counter.year').mean('time_counter').compute()
lin_gai_s_gtotal = esmtools.stats.linregress(gai_s_gtotal, dim="year")

print(lin_gai_s_gtotal.sel(parameter='slope').values)
print(lin_gai_s_gtotal.sel(parameter='pvalue') < 0.05 )# if True i significant

from scipy.stats import t
tinv = lambda p, df: abs(t.ppf(p/2, df))

ts = tinv(0.05, 13-2) # = 2.1788 # is the number of years for the ssh and ecco trends

print(f" slope (95%): {lin_gai_s_gtotal.sel(parameter='slope').values:.2f}  +/- {ts*lin_gai_s_gtotal.sel(parameter='stderr').values:.2f}")
```

```python
chunk_size = {"x":500,"y":500}
diri="/gpfswork/rech/cli/rcli002/eORCA025.L75/eORCA025.L75-I/"
mesh_hgr = xr.open_dataset(diri+'mesh_hgr.nc',chunks=chunk_size).squeeze()
tmask = mesh_hgr.tmask
e1t = mesh_hgr.e1t.fillna(0)
e2t = mesh_hgr.e2t.fillna(0)
# constructing the weight
bt = (e1t * e2t).load()# area of each cell

path = '/gpfswork/rech/cli/uor98hu/MYDATA/SSH/'
sshGAI =xr.open_dataset(path+'GAI/eORCA025.L75-IMHOTEP.GAI_y1980_2018.1m_sossheigCorrected.nc',chunks=chunk_size).sossheig
sshAI = xr.open_dataset(path+'AI/eORCA025.L75-IMHOTEP.AI_y1980_2018.1m_sossheigCorrected.nc', chunks=chunk_size).sossheig
sshS = xr.open_dataset(path+'S/eORCA025.L75-IMHOTEP.S_y1980_2018.1m_sossheigCorrected.nc', chunks=chunk_size).sossheig

# selecting the Beaufort Gyre box and doing the weighted mean over the box

maskBG = xr.open_dataset('/gpfswork/rech/cli/uor98hu/MYDATA/TRANSPORT_WP1/BGRmaskobs.nc', chunks={"x":500,"y":500}).BGRmask.squeeze()

btboxBG = bt.where(maskBG,drop=True)
tmaskboxBG = tmask.where(maskBG,drop=True)
oceanArea = (btboxBG*tmaskboxBG[0])  #tmask : 1 on ocean, 0 on land
totalOceanArea = oceanArea.sum()
WeightboxBG = (oceanArea/totalOceanArea) #.compute()

sshGAIboxBG =(sshGAI.where(maskBG,drop=True) * WeightboxBG).sum(('x','y'))
sshAIboxBG = (sshAI.where(maskBG,drop=True)  * WeightboxBG).sum(('x','y'))
sshSboxBG  = (sshS.where(maskBG,drop=True)   * WeightboxBG).sum(('x','y'))

diffWP1GAI_S = sshGAIboxBG - sshSboxBG
diffWP1GAI_AI = sshGAIboxBG - sshAIboxBG
diffWP1AI_S = sshAIboxBG -sshSboxBG

diffWP1GAI_Srllmean = diffWP1GAI_S.rolling(time_counter=12).mean()
diffWP1GAI_AIrllmean = diffWP1GAI_AI.rolling(time_counter=12).mean()
diffWP1AI_Srllmean = diffWP1AI_S.rolling(time_counter=12).mean()

print(str(np.round(xr.corr((GAI_Slist[0]), diffWP1GAI_Srllmean).values,2)))
print(str(np.round(xr.corr((GAI_AIlist[0]), diffWP1GAI_AIrllmean).values,2)))
print(str(np.round(xr.corr((AI_Slist[0]), diffWP1AI_Srllmean).values,2)))
```
