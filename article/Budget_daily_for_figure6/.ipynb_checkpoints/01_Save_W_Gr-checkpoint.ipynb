AIM : I used this notebook to compute Ug = Ugai - Uai (resp Sg and Vg) en 1d. The saving was unfortunately made using not wisely chosen variable and file names. therefore the files and variables were renamed afeterwise using ncrename. I addition, the scale factors of GAI e3u and e3v were added afterwards using ncks to the fils to reproduce typical gridU and gridV files used in the cdftool cdfvT. 

```python
import os,sys
import numpy as np
# xarray
import xarray as xr

from dask.distributed import Client
c = Client()
c
```

```python
prefix = "eORCA025.L75-IMHOTEP"
fo="1d" 
diridat="/gpfsstore/rech/cli/rcli002/eORCA025.L75/"+prefix+"."
diro = "/gpfsscratch/rech/cli/uor98hu/BILANS/1dDecomposition/"

chunk_size = {"time_counter":10} # {"x":500,"y":500}
```

```python
for year in np.arange(1982,2019):
    y1 = str(year)
    print(y1)
    # create directory for the year
    diroyear = y1+"-concat"
    diropath = os.path.join(diro, diroyear)
    os.mkdir(diropath)
    i = 0
    for month in ['01','02','03','04','05','06','07','08','09','10','11','12']: # 
        print("month",month)

        ######## ---------- load GAI
        nexp = "GAI"
        print(nexp)

        fileT = diridat+nexp+"-S/"+"/1d/"+y1+"-concat"+"/"+prefix+"."+nexp+"_y"+y1+"m"+month+"_1d_gridT.nc"
        fileU = diridat+nexp+"-S/"+"/1d/"+y1+"-concat"+"/"+prefix+"."+nexp+"_y"+y1+"m"+month+"_1d_gridU.nc"
        fileV = diridat+nexp+"-S/"+"/1d/"+y1+"-concat"+"/"+prefix+"."+nexp+"_y"+y1+"m"+month+"_1d_gridV.nc"
        
        T_ds = xr.open_dataset(fileT,chunks = chunk_size).squeeze()
        U_ds = xr.open_dataset(fileU,chunks = chunk_size).squeeze()
        V_ds = xr.open_dataset(fileV,chunks = chunk_size).squeeze()

        U_GAI = U_ds.vozocrtx
        V_GAI = V_ds.vomecrty
        S_GAI = T_ds.vosaline
        
    
        ######## ---------- load AI
        nexp = "AI"
        print(nexp)

        fileT = diridat+nexp+"-S/"+"/1d/"+y1+"-concat"+"/"+prefix+"."+nexp+"_y"+y1+"m"+month+"_1d_gridT.nc"
        fileU = diridat+nexp+"-S/"+"/1d/"+y1+"-concat"+"/"+prefix+"."+nexp+"_y"+y1+"m"+month+"_1d_gridU.nc"
        fileV = diridat+nexp+"-S/"+"/1d/"+y1+"-concat"+"/"+prefix+"."+nexp+"_y"+y1+"m"+month+"_1d_gridV.nc"
        
        T_ds = xr.open_dataset(fileT,chunks = chunk_size).squeeze()
        U_ds = xr.open_dataset(fileU,chunks = chunk_size).squeeze()
        V_ds = xr.open_dataset(fileV,chunks = chunk_size).squeeze()

        U_AI = U_ds.vozocrtx
        V_AI = V_ds.vomecrty
        S_AI = T_ds.vosaline
        
        ####### --------------- compute Greenland's terms
        U_G = U_GAI - U_AI
        V_G = V_GAI - V_AI
        S_G = S_GAI - S_AI
        
        ####### ------------- save terms
        dsU_G = U_G.to_dataset(name = 'U_G')
        dsV_G = V_G.to_dataset(name = 'V_G')
        dsS_G = S_G.to_dataset(name = 'S_G')
        print("saving")
        print("saving U_G")

        dsU_G.to_netcdf(diropath+"/U_G_y"+y1+"m"+month+"_1d.nc", mode='w')
        print("saving V_G")
        dsV_G.to_netcdf(diropath+"/V_G_y"+y1+"m"+month+"_1d.nc", mode='w')
        print("saving S_G")
        dsS_G.to_netcdf(diropath+"/S_G_y"+y1+"m"+month+"_1d.nc", mode='w')
```
